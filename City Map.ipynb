```python
import xml.etree.ElementTree as ET
import re
import pprint

tree=ET.parse('C:/Users/ADMIN/Desktop/Advanced_P1/suzhou.osm')
root=tree.getroot()
```

```python
#Iterate all the tags in the osm file, to have an overview of the number of tags and sub-tags in the file

tagset=set()
tags={}
taglist=[]
for title in root.iter():
    taglist.append(title.tag)
    tagset.add(title.tag)
for tag in tagset:
    tags[tag]=taglist.count(tag)
print tags
```

```python
'''
Define the function to iterate the file to get List for the specific building.
The list should include the name and English name, classification, address and postcode.
'''

def getlist(listname,feature):
    listname=[]
    for way in root.findall('./way'):
        tags=way.findall('tag')
        lclass=None
        ltype=None
        name_en=None
        name=None
        addr_street=None
        addr_housenumber=None
        addr_postcode=None
        for tag in tags:
            if tag.attrib['k']==feature:
                lclass=tag.attrib['v']
                ltype=tag.attrib['k']
            if tag.attrib['k']=="name:en":
                name_en=tag.attrib['v']
            if tag.attrib['k']=="name":
                name=tag.attrib['v']
            if tag.attrib['k']=="addr:street":
                addr_street=tag.attrib['v']
            if tag.attrib['k']=="addr:housenumber":
                addr_housenumber=tag.attrib['v']
            if tag.attrib['k']=="addr:postcode":
                addr_postcode=tag.attrib['v']
            if name is not None and ltype==feature:
                data = {
                    "name":name,
                    "name_en":name_en,
                    "lclass":lclass,
                    "addr_street":addr_street,
                    "addr_housenumber":addr_housenumber,
                    "addr_postcode":addr_postcode
                }
                listname.append(data)
    return listname
```

```python
#Call the previous function to get the list for roads, tourisms, amenities and shops

roads=getlist("roadname","highway")
tourisms=getlist("tourismname","tourism")
amenities=getlist("amenityname","amenity")
shops=getlist("shopname","shop")
```

# Solve problems of Over-Abbreviated

```python
def checkissues(filename,field):
    nameset=set()
    wordlist=[]
    wordcount={}
    for item in filename:
        if item[field] is not None:
            wordlist.extend(item[field].split())
    for item in wordlist:
        nameset.add(item)
    for item in nameset:
        wordcount[item]=wordlist.count(item)
    pprint.pprint(wordcount)
```

```python
#Get all the words in the English name of Roads and other lists,see whether their name is standard
print "Roads:"
checkissues(roads,'name_en')

print "Tourisms:"
checkissues(tourisms,'name_en')

print "Amenities:"
checkissues(amenities,'name_en')

print "Shops:"
checkissues(shops,'name_en')



#经过运行，我们发现仅roads列表中的地名需要修正
```

```python
#Modify the English name of the roads

expected=["Road","Street","West","Middle","North"]
mapping={
        "Lu":"Road",
        "Mid":"Middle",
        "N.":"North",
        "Rd":"Road",
        "St":"Street",
        "Str":"Street",
        "W.":"West",
        "road":"Road"
}
```

```python
#Define the function to update the name to its modified version

def update_name(name,mapping):
    for (key,value) in mapping.items():
        if re.search(key,name):
            name_words=name.split()
            i=0
            for item in name_words:
                if item in mapping.keys():
                    name_words[i]=value
                i+=1
            a=" "
            name=a.join(name_words)
    return name
```

```python
#Call the update_name function to update the English name in list Roads

for item in roads:
    if item['name_en'] is not None:
        item['name_en']=update_name(item['name_en'],mapping)
pprint.pprint(roads[0:10])
```

```python
def updatephone(number):
    if len(number)==6 or len(number)==10 or len(number)==11:
        return number
    else:
        striped_number=filter(str.isdigit,number.replace(' ',''))
        return striped_number[-11:-8]+" "+striped_number[-8:]
```

```python
#Define the function createDictCSV to load the list to csv file

import csv

def createDictCSV(filename,outfile):
    fieldnames=['lclass','name_en','name','addr_street','addr_housenumber','addr_postcode']
    with open(outfile,'w') as f:
        f_csv = csv.DictWriter(f, fieldnames)
        f_csv.writeheader()
        f_csv.writerows(filename)
```

```python
#Set the format for the node,node_tags,way,way_nodes and way_tags

SCHEMA = {
    'node': {
        'type': 'dict',
        'schema': {
            'id': {'required': True, 'type': 'integer', 'coerce': int},
            'lat': {'required': True, 'type': 'float', 'coerce': float},
            'lon': {'required': True, 'type': 'float', 'coerce': float},
            'user': {'required': True, 'type': 'string'},
            'uid': {'required': True, 'type': 'integer', 'coerce': int},
            'version': {'required': True, 'type': 'string'},
            'changeset': {'required': True, 'type': 'integer', 'coerce': int},
            'timestamp': {'required': True, 'type': 'string'}
        }
    },
    'node_tags': {
        'type': 'list',
        'schema': {
            'type': 'dict',
            'schema': {
                'id': {'required': True, 'type': 'integer', 'coerce': int},
                'key': {'required': True, 'type': 'string'},
                'value': {'required': True, 'type': 'string'},
                'type': {'required': True, 'type': 'string'}
            }
        }
    },
    'way': {
        'type': 'dict',
        'schema': {
            'id': {'required': True, 'type': 'integer', 'coerce': int},
            'user': {'required': True, 'type': 'string'},
            'uid': {'required': True, 'type': 'integer', 'coerce': int},
            'version': {'required': True, 'type': 'string'},
            'changeset': {'required': True, 'type': 'integer', 'coerce': int},
            'timestamp': {'required': True, 'type': 'string'}
        }
    },
    'way_nodes': {
        'type': 'list',
        'schema': {
            'type': 'dict',
            'schema': {
                'id': {'required': True, 'type': 'integer', 'coerce': int},
                'node_id': {'required': True, 'type': 'integer', 'coerce': int},
                'position': {'required': True, 'type': 'integer', 'coerce': int}
            }
        }
    },
    'way_tags': {
        'type': 'list',
        'schema': {
            'type': 'dict',
            'schema': {
                'id': {'required': True, 'type': 'integer', 'coerce': int},
                'key': {'required': True, 'type': 'string'},
                'value': {'required': True, 'type': 'string'},
                'type': {'required': True, 'type': 'string'}
            }
        }
    }
}
```

```python
'''
This file is to load the csv file from the osm.
We have set the format for the five files to be created through SCHEMA.
'''

import csv
import codecs

import cerberus

OSM_PATH = "C:/Users/ADMIN/Desktop/Advanced_P1/suzhou.osm"

NODES_PATH = "nodes.csv"
NODE_TAGS_PATH = "nodes_tags.csv"
WAYS_PATH = "ways.csv"
WAY_NODES_PATH = "ways_nodes.csv"
WAY_TAGS_PATH = "ways_tags.csv"

LOWER_COLON = re.compile(r'^([a-z]|_)+:([a-z]|_)+')
PROBLEMCHARS = re.compile(r'[=\+/&<>;\'"\?%#$@\,\. \t\r\n]')


# Set the fields of the list for the files to be created
NODE_FIELDS = ['id', 'lat', 'lon', 'user', 'uid', 'version', 'changeset', 'timestamp']
NODE_TAGS_FIELDS = ['id', 'key', 'value', 'type']
WAY_FIELDS = ['id', 'user', 'uid', 'version', 'changeset', 'timestamp']
WAY_TAGS_FIELDS = ['id', 'key', 'value', 'type']
WAY_NODES_FIELDS = ['id', 'node_id', 'position']

#Take value from the osm file, and return as the distionary
def shape_element(element, node_attr_fields=NODE_FIELDS, way_attr_fields=WAY_FIELDS,
                  problem_chars=PROBLEMCHARS, default_tag_type='regular'):
    """Clean and shape node or way XML element to Python dict"""

    node_attribs = {}
    way_attribs = {}
    way_nodes = []
    tags = []  # Handle secondary tags the same way for both node and way elements

    # save the tags in the list
    if element.tag == 'node':
        for i in NODE_FIELDS:
            node_attribs[i] = element.attrib[i]
        for child in element:
            node_tags = {}
            colon = re.search(LOWER_COLON,child.attrib['k'])
            prob = re.search(PROBLEMCHARS,child.attrib['k'])
            if colon:
                node_tags['id'] = element.attrib['id']
                node_tags['key'] = child.attrib['k'].split(':')[1]
                if child.attrib['k']=='name:en':
                    node_tags['value'] = update_name(child.attrib['v'],mapping)
                else:
                    node_tags['value'] = child.attrib['v']
                node_tags['type'] = child.attrib['k'].split(':')[0]
            elif prob:
                continue
            else:
                node_tags['key'] = child.attrib['k'].split(':')[0]
                node_tags['type'] = 'regular'
                node_tags['id']=element.attrib['id']
                if child.attrib['k']=='phone':
                    node_tags['value']=updatephone(child.attrib['v'])
                else:
                    node_tags['value']=child.attrib['v']
            tags.append(node_tags) 
        return {'node': node_attribs, 'node_tags': tags}

    elif element.tag == 'way':
        count = 0
        for i in WAY_FIELDS:
            way_attribs[i] = element.attrib[i]
        for child in element:
            way_tags = {}
            way_nodes_v = {}
            if child.tag == 'tag':
                colon = LOWER_COLON.search(child.attrib['k'])
                prob = PROBLEMCHARS.search(child.attrib['k'])
                if colon:
                    way_tags['id'] = element.attrib['id']
                    way_tags['key'] = ':'.join(child.attrib['k'].split(':')[1:])
                    if child.attrib['k']=='name:en':
                        way_tags['value'] = update_name(child.attrib['v'],mapping)
                    else:
                        way_tags['value'] = child.attrib['v']
                    way_tags['type'] = child.attrib['k'].split(':')[0]
                    print way_tags
                elif prob:
                    continue
                else:
                    way_tags['key'] = child.attrib['k']
                    way_tags['type'] = 'regular'
                    way_tags['id'] = element.attrib['id']
                    if child.attrib['k']=='phone':
                        node_tags['value']=updatephone(child.attrib['v'])
                    else:
                        way_tags['value'] = child.attrib['v']
                tags.append(way_tags)
        
            if child.tag == 'nd':
                way_nodes_v['id'] = element.attrib['id']
                way_nodes_v['node_id'] = child.attrib['ref']
                way_nodes_v['position'] = count
                count += 1
                way_nodes.append(way_nodes_v)
    
        return {'way': way_attribs, 'way_nodes': way_nodes, 'way_tags': tags}


# ================================================== #
#               Helper Functions                     #
# ================================================== #
def get_element(osm_file, tags=('node', 'way', 'relation')):
    """Yield element if it is the right type of tag"""

    context = ET.iterparse(osm_file, events=('start', 'end'))
    _, root = next(context)
    for event, elem in context:
        if event == 'end' and elem.tag in tags:
            yield elem
            root.clear()


def validate_element(element, validator, schema=SCHEMA):
    """Raise ValidationError if element does not match schema"""
    if validator.validate(element, schema) is not True:
        field, errors = next(validator.errors.iteritems())
        message_string = "\nElement of type '{0}' has the following errors:\n{1}"
        error_string = pprint.pformat(errors)
        
        raise Exception(message_string.format(field, error_string))


class UnicodeDictWriter(csv.DictWriter, object):
    """Extend csv.DictWriter to handle Unicode input"""

    def writerow(self, row):
        super(UnicodeDictWriter, self).writerow({
            k: (v.encode('utf-8') if isinstance(v, unicode) else v) for k, v in row.iteritems()
        })

    def writerows(self, rows):
        for row in rows:
            self.writerow(row)


# ================================================== #
#               Main Function                        #
# ================================================== #
def process_map(file_in, validate):
    """Iteratively process each XML element and write to csv(s)"""

    with codecs.open(NODES_PATH, 'w') as nodes_file, \
         codecs.open(NODE_TAGS_PATH, 'w') as nodes_tags_file, \
         codecs.open(WAYS_PATH, 'w') as ways_file, \
         codecs.open(WAY_NODES_PATH, 'w') as way_nodes_file, \
         codecs.open(WAY_TAGS_PATH, 'w') as way_tags_file:

        nodes_writer = UnicodeDictWriter(nodes_file, NODE_FIELDS)
        node_tags_writer = UnicodeDictWriter(nodes_tags_file, NODE_TAGS_FIELDS)
        ways_writer = UnicodeDictWriter(ways_file, WAY_FIELDS)
        way_nodes_writer = UnicodeDictWriter(way_nodes_file, WAY_NODES_FIELDS)
        way_tags_writer = UnicodeDictWriter(way_tags_file, WAY_TAGS_FIELDS)

        nodes_writer.writeheader()
        node_tags_writer.writeheader()
        ways_writer.writeheader()
        way_nodes_writer.writeheader()
        way_tags_writer.writeheader()

        validator = cerberus.Validator()

        for element in get_element(file_in, tags=('node', 'way')):
            el = shape_element(element)
            if el:
                if validate is True:
                    validate_element(el, validator)

                if element.tag == 'node':
                    nodes_writer.writerow(el['node'])
                    node_tags_writer.writerows(el['node_tags'])
                elif element.tag == 'way':
                    ways_writer.writerow(el['way'])
                    way_nodes_writer.writerows(el['way_nodes'])
                    way_tags_writer.writerows(el['way_tags'])


process_map(OSM_PATH, validate=True)
```
